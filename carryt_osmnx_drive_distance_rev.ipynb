```python
import osmnx as ox
import networkx as nx
import os

%matplotlib inline
ox.__version__

import warnings
warnings.filterwarnings('ignore')

ox.config(log_console=True, use_cache=True)
```

# Mode = Drive

```python
STREETGRAPH_FILENAME = '/content/drive/MyDrive/omdena/carryt/carryt.graphml'
city = ['Bogot√°, Colombia'] 

if (not os.path.isfile(STREETGRAPH_FILENAME) ):
    G = ox.graph_from_place(city, network_type='drive',retain_all=True)
    ox.save_graphml(G, STREETGRAPH_FILENAME)
else:
    G = ox.load_graphml(STREETGRAPH_FILENAME)

startpoint = (4.67215552, -74.150175499) 
endpoint = (4.5809172, -74.1975297) 

# find the nearest node to the start location
node_start = ox.distance.get_nearest_node(G, startpoint)

# find the nearest node to the end location
node_end = ox.distance.get_nearest_node(G, endpoint)

print(node_start,node_end)
```

## Shortest route based on travel time

```python
# find shortest path length based on distance 
optimizer = 'travel_time'        
travel_time = nx.shortest_path_length(G, node_start, node_end, weight=optimizer)

print (f'Travel time(mins) : {travel_time}')
```

```python
route = nx.shortest_path(G, node_start, node_end, weight=optimizer)
```

```python
ox.plot_route_folium(G, route, route_color='blue',tiles='openstreetmap' ,weight=5)
```

## Shortest route based on distance

```python
# find shortest path length based on distance 
optimizer = 'length'        
distance = nx.shortest_path_length(G, node_start, node_end, weight=optimizer)

print (f'Driving distance in meters : {distance}')
```

```python
route = nx.shortest_path(G, node_start, node_end, weight='length')
```

```python
ox.plot_route_folium(G, route, route_color='blue',tiles='openstreetmap' ,weight=5)
```
